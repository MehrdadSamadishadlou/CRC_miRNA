# Expression profiles of selected miRNAs

This notebook generates the violin and boxplot figure for the 24 candidate miRNAs using the training data.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## File locations

```python
project_root = Path("..")
data_dir = project_root / "data" / "processed"
figure_dir = project_root / "results" / "figures"

figure_dir.mkdir(parents=True, exist_ok=True)
```

## Load and reshape the data

```python
train_data = pd.read_table(
    data_dir / "train_exp_selected_mirs.txt",
    sep=" "
)

train_data.loc[
    train_data["stage"] != "Healthy",
    "stage"
] = "CRC"

print(train_data["stage"].value_counts())
```

```python
long_expression = train_data.melt(id_vars="stage")
long_expression["value"] = long_expression["value"].astype(float)

long_expression.head()
```

## Plot selected miRNA expression

```python
violin_colors = [
    (0.5529411764705883, 0.8980392156862745, 0.6313725490196078),
    (1.0, 0.6235294117647059, 0.6078431372549019)
]

box_colors = [
    (0.10196078431372549, 0.788235294117647, 0.2196078431372549),
    (0.9098039215686274, 0.0, 0.043137254901960784)
]
```

```python
plt.figure(figsize=(20, 8))
sns.set_style("white")
plt.rcParams["font.family"] = "Times"

axis = sns.violinplot(
    x="variable",
    y="value",
    data=long_expression,
    hue="stage",
    dodge=True,
    saturation=0.9,
    palette=violin_colors,
    width=0.6,
    inner=None,
    scale="width",
    gap=0.1,
    split=True
)

sns.boxplot(
    x="variable",
    y="value",
    data=long_expression,
    hue="stage",
    showfliers=False,
    width=0.45,
    palette=box_colors,
    boxprops={"zorder": 1, "alpha": 0.7},
    ax=axis
)

handles, labels = axis.get_legend_handles_labels()
axis.legend(
    handles[2:],
    labels[2:],
    loc="upper right",
    prop={"size": 17}
)

plt.ylabel("Expression Value (log2)", fontsize=19)
plt.xlabel("")
plt.tick_params(axis="x", which="major", labelsize=17)
axis.set_xticklabels(
    axis.get_xticklabels(),
    rotation=35,
    ha="right"
)
plt.tick_params(axis="y", which="major", labelsize=17)

plt.savefig(
    figure_dir / "selected_miRNA_expression.pdf",
    bbox_inches="tight"
)
plt.show()
```
