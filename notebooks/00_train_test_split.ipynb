# Training and hold-out test split

This notebook creates the stratified 75:25 training and hold-out test sets used in the study.


## Import packages


```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

## Define input and output files


```python
expression_file = "../data/processed/Expression_all.csv"
label_file = "../data/processed/exp_all_normal.txt"

train_output_file = "../data/processed/train_set.csv"
test_output_file = "../data/processed/test_set.csv"
```

## Load the expression matrix


```python
expression_data = pd.read_csv(
    expression_file,
    low_memory=False,
    index_col="Unnamed: 0"
)
```

## Load the sample labels


```python
expression_with_labels = pd.read_table(
    label_file,
    sep=" ",
    low_memory=False
)

sample_labels = expression_with_labels.iloc[-1, :].copy()
```

## Arrange samples as rows and miRNAs as columns


```python
sample_expression = expression_data.astype("float64").T
```

## Create the stratified training and hold-out test sets


```python
X_train, X_test, y_train, y_test = train_test_split(
    sample_expression,
    sample_labels,
    test_size=0.25,
    random_state=1,
    stratify=sample_labels
)
```

## Add the class labels


```python
train_set = pd.concat([X_train, y_train], axis=1)
test_set = pd.concat([X_test, y_test], axis=1)
```

## Save the split datasets


```python
train_set.to_csv(train_output_file)
test_set.to_csv(test_output_file)
```
